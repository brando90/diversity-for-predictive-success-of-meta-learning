```python
import matplotlib.pyplot as plt
def utri(x):
    arr = []
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            if(i < j):
                arr += [x[i][j]]
    return arr

def plothm(hm,experiment):
    hm = utri(hm)
    
    fig=plt.figure(dpi=200)
    fig.patch.set_facecolor('white')

    n, bins, patches = plt.hist(hm, 10, facecolor='b', alpha=0.75,edgecolor = "black")

    plt.xlabel('Cosine Distance between Task Pairs')
    plt.ylabel('Frequency')
    plt.title('Distribution of Task2Vec Distances, ' + experiment)
    #plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
    #plt.xlim(40, 160)
    #plt.ylim(0, 0.03)
    #plt.grid(True)
    plt.show()
    #sns.title
    #sns.histplot(data=hm001)


for folder, exp in zip(folders, experiment_name):
    res = torch.load(folder + '/' + 'results.pt')['distance_matrix']
    plothm(res,exp)


```
